```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_orig = np.loadtxt('1_orig.txt')
data_pred = np.loadtxt('pos_txt.txt')
```

```python
orig = data_orig.T
pred = data_pred.T
orig_x,pred_x = orig[0,0:328],pred[0,:]
plt.scatter(orig_x,pred_x,c="r")
plt.xlabel("Orig_x_pos")
plt.ylabel("Pred_x_pos")
```

```python
x_time = np.arange(328)
```

```python
fig,ax=plt.subplots()
ax.plot(x_time, orig_x, marker=".")
ax.plot(x_time, pred_x, marker=".")
ax.set_xlabel("Time")
ax.set_ylabel("X_pos")
ax.legend(['Orig','Pred'])
plt.show()
```

```python
orig_y = data_orig[:,1]
orig_y,pred_y = orig_y[0:328],data_pred[:,1]
plt.scatter(pred_y,orig_y,c = 'g')
plt.xlabel("Orig_y_pos")
plt.ylabel("Pred_y_pos")
```

```python
fig,ax=plt.subplots()
ax.plot(x_time, orig_y, marker=".")
ax.plot(x_time, pred_y, marker=".")
ax.set_xlabel("Time")
ax.set_ylabel("Y_pos")
ax.legend(['Orig','Pred'])
plt.show()
```
